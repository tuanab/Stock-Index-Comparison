```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
Tickers = ['^GSPC', '^IXIC', '^GDAXI', '^FTSE']
index_data = pd.DataFrame()
for t in Tickers:
    index_data[t] = wb.DataReader(t, data_source = 'yahoo', start = '1997-1-1')['Adj Close']
```

```python
index_data.head()
```

```python
index_data.info()
```

```python
returns = (index_data / index_data.shift(1)) -1
returns.tail()
```

```python
average_annual_return = returns.mean() * 250
average_annual_return
```

```python
normalized_returns = (index_data / index_data.loc['1997-01-02'] * 100)
normalized_returns.tail()
```

```python
normalized_returns.plot(figsize = (18, 5))
plt.show()
```

```python
index_data.plot(figsize = (18, 5))     # index price charts
plt.show()
```
